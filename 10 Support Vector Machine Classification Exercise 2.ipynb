## Train SVM classifier using sklearn breast cancer dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer

import warnings
warnings.filterwarnings('ignore')
```

```python
cancer = load_breast_cancer()
```

```python
dir(cancer)
```

```python
cancer.data
```

```python
cancer.feature_names
```

```python
cancer.target
```

```python
cancer.target_names
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df.head()
```

```python
df['target'] = cancer.target
df.head()
```

```python
df['target_names'] = df.target.apply(lambda x: cancer.target_names[x])
df.head()
```

## Visualize the relationship between our features

```python
import seaborn as sns
sns.pairplot(df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'] )
```

Note:
1.0 (Orange) = Benign (No Cancer)
0.0 (Blue) = Malignant (Cancer)

## How many Benign and Malignant do we have in our dataset?

```python
df.groupby('target_names').describe()
```

```python
df['target'].value_counts()
```

```python
sns.countplot(df['target'], label = "Count")
```

```python
X = df.drop(['target', 'target_names'], axis = 'columns')
X.head()
```

```python
y = df['target']
y.head()
```

## Create the training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Support Vector Machine (SVM) Model

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

## The confusion matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_predicted)
confusion_matrix
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))
sn.heatmap(confusion_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
